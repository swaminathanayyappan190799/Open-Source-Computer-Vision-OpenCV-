This notebook includes reading of images using PIL (Python Imaging Library), and converting those images into an array with the use of numpy . Finally that image are visualized based on every individual channels (RGB). 

## Importing necessary libraries

```python
import matplotlib.pyplot as plt #For visualization
import numpy as np #For handling arrays
from PIL import Image #For handling images
```

## Reading input image from google drive

```python
#Image reading using PIL
image=Image.open('/content/drive/MyDrive/cosmos-flower.jpg')
```

```python
#displaying the input image.
image
```

## Image to array conversion

```python
#Converting the input image into numpy array.
img_array=np.asarray(image)
```

```python
#checking the type of the input image after conversion.
type(img_array)
```

```python
#Plotting the converted image array using matplotlib's imshow.
plt.imshow(img_array)
```

```python
#Gives the shape of the input image array
img_array.shape
```

```python
#Input array image before rescaling.
img_array
```

## Rescaling

```python
#Rescaling the images in order to bring it on a common scale.
arr_copy=img_array.copy()/255
```

## Making multiple copies of input array image

```python
copy1=img_array.copy()/255 #For Red Channel Image Visualization (Rescaled)
copy2=img_array.copy()/255 #For Green Channel Image Visualization (Rescaled)
copy3=img_array.copy()/255 #For Blue Channel Image Visualization (Rescaled)
```

```python
#Nullifying rest of the channels to visualize red channel.
copy1[:,:,1]=0
copy1[:,:,2]=0
#Nullifying rest of the channels to visualize green channel.
copy2[:,:,0]=0
copy2[:,:,2]=0
#Nullifying rest of the channels to visualize blue channel.
copy3[:,:,0]=0
copy3[:,:,1]=0
```

## Visualizing images with seperate channels

```python
#Mentioning the size of the figure
plt.figure(figsize=(20,10))
#Creating the rescaled input image as the first plot in the subplot. 
plt.subplot(2,3,2)
plt.title('Rescaled input image')
plt.xlabel('Width')
plt.ylabel('Height')
plt.imshow(img_array)
#Creating the rescaled red channel image as the second plot in the subplot.
plt.subplot(2,3,4)
plt.title('Red Channel')
plt.xlabel('Width')
plt.ylabel('Height')
plt.imshow(copy1)
#Creating the rescaled green channel image as the third plot in the subplot.
plt.subplot(2,3,5)
plt.title('Green Channel')
plt.xlabel('Width')
plt.ylabel('Height')
plt.imshow(copy2)
#Creating the rescaled blue channel image as the fourth plot in the subplot.
plt.subplot(2,3,6)
plt.title('Blue Channel')
plt.xlabel('Width')
plt.ylabel('Height')
plt.imshow(copy3)
plt.show()
```
