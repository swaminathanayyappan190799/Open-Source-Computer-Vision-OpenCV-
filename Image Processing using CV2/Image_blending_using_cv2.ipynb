```python
#Converting the watermark color format from BGR to RGB.
rgb_watermark=cv2.cvtColor(watermark,cv2.COLOR_BGR2RGB)
```

## **Image blending**

The process of pasting one image over another image is known as Image blending.Whereas the pixel for the blended image can be determined by the following equation.


**pixel_for_blended_image = (alpha * pixel for image1) + (beta * pixel for image2) + gamma** 


Note : For performing image blending both images should be of same size.

```python
#checking shape of the flower image which is in RGB color format that is our imgae1 for image blending.
rgb_image.shape
```

```python
#checking shape of the watermark image which is in RGB color format that is our image2 for image blending.
rgb_watermark.shape
```

```python
#resizing both images to have a same size.
rgb_watermark=cv2.resize(rgb_watermark,(300,300))
rgb_image=cv2.resize(rgb_image,(300,300))
```

```python
#Checking the size of the flower image after reshaping.
rgb_image.shape
```

```python
#Checking the size of the watermark image after reshaping.
rgb_watermark.shape
```

```python
#Blending images using addWeighted function in cv2.
blended_img=cv2.addWeighted(src1=rgb_image,alpha=1.0,src2=rgb_watermark,beta=0.2,gamma=0)
```

```python
plt.figure(figsize=(8,8)) #Mentioning the size for the blended image plot.
plt.imshow(blended_img) #Visualizing the blended image.
plt.title('Blended image') #Giving title to the plot.
plt.ylabel('Height of the image') #Giving Xlabel to the plot.
plt.xlabel('Width of the image') #Giving Ylabel to the plot.
```

## **Overlaying of images**

The process of pasting an image over another image is known as overlaying of image. In simple terms rather than pasting images of same resolution , we can do blending of images that has different resolutions.

```python
flower_img=cv2.imread('/content/drive/MyDrive/cosmos-flower.jpg') #Reading an input image from google cloud.
flower_img=cv2.cvtColor(flower_img,cv2.COLOR_BGR2RGB) #Converting the color format from BGR to RGB.
```

```python
#Gives out the shape of the input flower image.
flower_img.shape
```

```python
#Gives out the shape of the resized watermark image.
rgb_watermark.shape
```

```python
X_start,y_start=(0,0) #Gives the starting co-ordinates for the image.
img1=flower_img #Making a copy of the input flower image.
img2=rgb_watermark #Making a copy of the resized watermark image.
X_end=X_start+img2.shape[1] #Gives the height end co-ordinates for the image.  
y_end=y_start+img2.shape[0] #Gives the width end co-ordinates for the image.
```

```python
#Overlaying of img2 over img1.
img1[y_start:y_end,X_start:X_end]=img2
```

```python
plt.figure(figsize=(10,8))
plt.imshow(img1) #Visualizing the overlayed image.
plt.title('Overlayed Image') #Giving title to the plot.
plt.xlabel('Width of the overlayed image') #Giving name to the Xaxis of the plot.
plt.ylabel('Height of the overlayed image') #Giving name to the Yaxis of the plot.
```
