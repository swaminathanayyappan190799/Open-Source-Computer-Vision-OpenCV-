This notebook explains of creating basic shapes (square,rectangle,triangle,circle) using cv2 in python.

## Importing necessary libraries

```python
import cv2 #Importing cv2
import matplotlib.pyplot as plt #For visualizing the array
import numpy as np #For handling arrays in python
```

## Creating an custom background for drawing shapes

```python
#Will creates an black background using numpy arrays which was used to draw basic shapes.
bg_img=np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
#Visualizing the plain black background image.
plt.figure(figsize=(10,10))
plt.imshow(bg_img)
plt.title('Background image for drawing shapes')
plt.xlabel('Width of the background image')
plt.ylabel('Height of the background image')
plt.show()
```

## Drawing a square 

```python
#Draws an square using the rectangle method in cv2 and visualizing it.
cv2.rectangle(bg_img,pt1=(400,100),pt2=(500,0),color=(120, 222, 18),thickness=10)
plt.figure(figsize=(10,10)) 
plt.imshow(bg_img)
plt.title('Drawing Square using cv2')
plt.xlabel('Width of the image')
plt.ylabel('Height of the image')
plt.show()
```

## Drawing a circle

```python
#Drawing a circle using cv2 on the background image using cv2 and visualizing it.
cv2.circle(bg_img,center=(250,250),radius=80,color=(18, 198, 222),thickness=10)
plt.figure(figsize=(10,10)) 
plt.imshow(bg_img)
plt.title('Drawing Circle using cv2')
plt.xlabel('Width of the image')
plt.ylabel('Height of the image')
plt.show()
```

## Drawing a rectangle

```python
#Drawing an rectangle using cv2 and visualizing it.
cv2.rectangle(bg_img,pt1=(150,200),pt2=(10,0),color=(72, 18, 222),thickness=-1) #The rectangle will get entirely filled out since it's thickness- is -1.
plt.figure(figsize=(10,10)) 
plt.imshow(bg_img)
plt.title('Drawing rectangle using cv2')
plt.xlabel('Width of the image')
plt.ylabel('Height of the image')
plt.show()
```

## Drawing a line 

```python
#Drawing a line using cv2 and visualizing it.
cv2.line(bg_img,pt1=(50,450),pt2=(350,450),color=(255, 117, 230),thickness=10)
plt.figure(figsize=(10,10)) 
plt.imshow(bg_img)
plt.title('Drawing line using cv2')
plt.xlabel('Width of the image')
plt.ylabel('Height of the image')
plt.show()
```

## Creating an custom triangle using cv2

```python
#creating an background for triangle shape drawing.
cust_tri_img=np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
#Drawing an custom triangle using cv2 with the use of line.
cv2.line(cust_tri_img,pt1=(150,400),pt2=(350,400),color=(255, 255, 168),thickness=8)
cv2.line(cust_tri_img,pt1=(150,400),pt2=(250,200),color=(255, 255, 168),thickness=8)
cv2.line(cust_tri_img,pt1=(250,200),pt2=(350,400),color=(255, 255, 168),thickness=8)
plt.figure(figsize=(10,10)) 
plt.imshow(cust_tri_img)
plt.title('Drawing triangle using cv2')
plt.xlabel('Width of the image')
plt.ylabel('Height of the image')
plt.show()
```

## Creating an backgroung for drawing an man waling over a staircase

```python
#Crating a background for drawing a man using cv2.
man_img=np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
#Drawing staircase using line in cv2.
cv2.line(man_img,pt1=(0,500),pt2=(0,450),color=(217, 168, 255),thickness=12)
cv2.line(man_img,pt1=(0,450),pt2=(100,450),color=(217, 168, 255),thickness=12)
cv2.line(man_img,pt1=(100,450),pt2=(100,400),color=(217, 168, 255),thickness=12)
cv2.line(man_img,pt1=(100,400),pt2=(200,400),color=(217, 168, 255),thickness=12)
cv2.line(man_img,pt1=(200,400),pt2=(200,350),color=(217, 168, 255),thickness=12)
cv2.line(man_img,pt1=(200,350),pt2=(300,350),color=(217, 168, 255),thickness=12)
cv2.line(man_img,pt1=(300,350),pt2=(300,300),color=(217, 168, 255),thickness=12)
cv2.line(man_img,pt1=(300,300),pt2=(400,300),color=(217, 168, 255),thickness=12)
cv2.line(man_img,pt1=(0,450),pt2=(100,450),color=(217, 168, 255),thickness=12)
cv2.line(man_img,pt1=(400,300),pt2=(400,250),color=(217, 168, 255),thickness=12)
cv2.line(man_img,pt1=(400,250),pt2=(500,250),color=(217, 168, 255),thickness=12)
plt.figure(figsize=(10,10)) 
plt.imshow(man_img)
plt.title('Drawing staircase using cv2')
plt.xlabel('Width of the image')
plt.ylabel('Height of the image')
plt.show()
```

```python
cv2.line(man_img,pt1=(230,350),pt2=(250,280),color=(255,255,255),thickness=5)
cv2.line(man_img,pt1=(280,350),pt2=(250,280),color=(255,255,255),thickness=5)
cv2.line(man_img,pt1=(250,280),pt2=(250,220),color=(255,255,255),thickness=5)
cv2.circle(man_img,center=(250,200),radius=30,color=(255,255,255),thickness=-1)
cv2.line(man_img,pt1=(250,250),pt2=(220,250),color=(217, 35, 35),thickness=5)
cv2.line(man_img,pt1=(220,250),pt2=(180,280),color=(217, 35, 35),thickness=5)
cv2.line(man_img,pt1=(250,250),pt2=(290,250),color=(217, 35, 35),thickness=5)
cv2.line(man_img,pt1=(290,250),pt2=(310,200),color=(217, 35, 35),thickness=5)
plt.figure(figsize=(10,10)) 
plt.imshow(man_img)
plt.title('Drawing a man standing on a staircase using cv2')
plt.xlabel('Width of the image')
plt.ylabel('Height of the image')
plt.show()
```

## Inserting text into the image

```python
#Inserting text into the image using cv2.
cv2.putText(man_img,text='Hi!!',org=(0,100),fontScale=2,thickness=4,color=(37, 194, 160),fontFace=cv2.FONT_ITALIC)
plt.figure(figsize=(10,10)) 
plt.imshow(man_img)
plt.title('Inserting text using cv2')
plt.xlabel('Width of the image')
plt.ylabel('Height of the image')
plt.show()
```
