The notebook explains of reading images using openCV and converting their color format from Blue-Green-Red(BGR) to Red-Green-Blue(RGB) and also grayscale using opencv using python. 

## Importing Libraries

```python
import matplotlib.pyplot as plt #For visualizing the image.
import cv2 #OpenCV importing
```

## Importing an image using OpenCV

There are two cases which is involved in importing an input image using cv2

1.   If the path provided inside cv2.imread is a correct path then the type of the input image will be an "**numpy array**".  
2.   Else, if the path provided inside cv2.imread is an incorrect path then the type of the input image will be shown as **"NoneType"**.

Usually a valid input image will be readed as an numpy while reading it using cv2


**Case - 1 :** Importing an image with it's absolute correct path using cv2

```python
#Importing the image with correct path using cv2
inp_image=cv2.imread('/content/drive/MyDrive/cosmos-flower.jpg')
```

```python
#Checking the type of the input image.
type(inp_image)
```

**Case - 2 :** Importing an image with a incorrect path using cv2

```python
#Importing the image with incorrect path using cv2
incorrect_img_path=cv2.imread('/content/drive/MyDrive/cosmos.jpg')
```

```python
#Checking the type of the incorrect image
type(incorrect_img_path)
```

Since the image path was imported using a wrong path it displays the type as **"NoneType"**.

```python
#Visualizing the input image using matplotlib.
plt.imshow(inp_image)
plt.title('Input image using CV2')
plt.xlabel('Width of the image')
plt.ylabel('Height of the image')
plt.show()
```

We can notice that the imported image using cv2 was not properly visualized in the above plot , that's because matplotlib expects the input image to be in the RGB color format , whereas cv2 will reads an input image in the color format of BGR.

```python
#Gives the shape of the input image
inp_image.shape
```

## Converting BGR to RGB

```python
#Converting BGR color format to RGB color format.
recolored_img=cv2.cvtColor(inp_image,cv2.COLOR_BGR2RGB)
```

```python
#Visualizing the converted image using matplotlib
plt.imshow(recolored_img)
plt.title('Converted Input image using cv2')
plt.xlabel('Width of the image')
plt.ylabel('Height of the image')
plt.show()
```

## Reading a grayscale image using cv2

```python
#Reading an image in grayscale color format.
grayinp_img=cv2.imread('/content/drive/MyDrive/cosmos-flower.jpg',0)
```

```python
#Converting the color format BGR to RGB.
grayinp_img=cv2.cvtColor(grayinp_img,cv2.COLOR_BGR2RGB)
```

```python
#Visualizing the grayscale image using matplotlib.
plt.imshow(grayinp_img)
plt.title('Grayscale input image using cv2')
plt.xlabel('Width of the image')
plt.ylabel('Height of the image')
plt.show()
```

## Image resizing

```python
#Resizing the image using dimensions.
resized_img=cv2.resize(recolored_img,(1000,800))
```

```python
#Plotting the resized image.
plt.imshow(resized_img)
plt.title('Re-sized input image using cv2')
plt.xlabel('Width of the image')
plt.ylabel('Height of the image')
plt.show()
```

```python
#Checking the size of the resized image.
resized_img.shape
```
